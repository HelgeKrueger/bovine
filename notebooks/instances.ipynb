```python
from urllib.parse import urlparse
from bovine.clients.nodeinfo import fetch_nodeinfo

with open('urls') as url_file:
    lines = url_file.readlines()
domains = {urlparse(x).netloc for x in lines}
```

```python
import asyncio
import aiohttp
session = aiohttp.ClientSession()

async def nodeinfo(session, domain):
    return {
        "domain": domain,
        "nodeinfo": await fetch_nodeinfo(session, domain)
    }

result = await asyncio.gather(*[nodeinfo(session, domain) for domain in domains])
```

```python
def parse(entry):
    try:
        nodeinfo = entry["nodeinfo"]
        software = nodeinfo["software"]["name"] + " " + nodeinfo["software"]["version"][:3]
        active_month = nodeinfo["usage"]["users"]["activeMonth"]
        return {
            "domain": entry["domain"],
            "software": software,
            "active_month": active_month
        }
    except:
        return None
    
result2= [x for x in [parse(entry) for entry in result] if x]

from collections import defaultdict

users_by_software = defaultdict(int)
for entry in result2:
    if entry["active_month"]:
        users_by_software[entry["software"]] += entry["active_month"]
    
for a, b in sorted(users_by_software.items(), key=lambda x: -x[1]):
    print(f"{a:20}: {b}")
```

```python
async def add_mastodon_info(entry):
    try:
        instance = entry["domain"]
        entry["public_api"] = len(await (await session.get(f'https://{instance}/api/v1/timelines/public')).json()) > 1
        return entry
    except:
        entry["public_api"] = False
        return entry


mastodons = await asyncio.gather(*[add_mastodon_info(entry) for entry in mastodons])
```

```python
no_public_api = [x for x in mastodons if not x["public_api"]]
len(no_public_api)
```
